```python

from abc import ABCMeta, abstractmethod
import copy

 
class Componente(object):
	__metaclass__ = ABCMeta
	
	nombre = ""
	def __init__(self, nombre):
		self.nombre = nombre
	
	
	@abstractmethod
	def imprimaPreorden(self,profundidad=0):
		pass
	
	def __str__(self):
		return self.nombre

class Carpeta (Componente):
	hijos = []

	def __init__(self, nombre):
		self.hijos = []
		Componente.__init__(self, nombre)
	
	def agregar(self, h):
		self.hijos.append(h)
	
	def imprimaPreorden (self, profundidad=0):
		print (str("\t"*profundidad) + "" + self.nombre)
		for h in self.hijos:
			h.imprimaPreorden(profundidad + 1)
	

class Archivo(Componente):

	def imprimaPreorden(self, profundidad=0):
		print (str("\t"*profundidad) + "" + self.nombre)
		
	




raiz = Carpeta("Carpeta raíz")
raiz.agregar(Archivo("Archivo 1"))
raiz.agregar(Archivo("Archivo 2"))

carpeta1 = Carpeta("Carpeta A")
carpeta1.agregar(Archivo("Archivo A1"))
carpeta1.agregar(Archivo("Archivo A2"))

carpeta3 = Carpeta("Carpeta AA")
carpeta3.agregar(Archivo("Archivo AA1"))
carpeta3.agregar(Archivo("Archivo AA2"))
carpeta3.agregar(Archivo("Archivo AA3"))
carpeta3.agregar(Archivo("Archivo AA4"))


carpeta4 = Carpeta("Carpeta AAA")
carpeta4.agregar(Archivo("Archivo AAA1"))
carpeta4.agregar(Archivo("Archivo AAA2"))
carpeta4.agregar(Archivo("Archivo AAA3"))
carpeta4.agregar(Archivo("Archivo AAA4"))
carpeta4.agregar(Archivo("Archivo AAA5"))
carpeta3.agregar(carpeta4)



carpeta3.agregar(Archivo("Archivo AA5"))
carpeta3.agregar(Archivo("Archivo AA6"))
carpeta3.agregar(Archivo("Archivo AA7"))
carpeta1.agregar(carpeta3)

carpeta1.agregar(Archivo("Archivo A3"))
carpeta1.agregar(Archivo("Archivo A4"))
carpeta1.agregar(Archivo("Archivo A5"))
carpeta1.agregar(Archivo("Archivo A6"))
carpeta1.agregar(Archivo("Archivo A7"))
raiz.agregar(carpeta1)

carpeta2 = Carpeta("Carpeta B")
carpeta2.agregar(Archivo("Archivo B1"))

carpeta5 = Carpeta("Carpeta BA")
carpeta5.agregar(Archivo("Archivo BA1"))
carpeta5.agregar(Archivo("Archivo BA2"))
carpeta5.agregar(Archivo("Archivo BA3"))
carpeta5.agregar(Archivo("Archivo BA4"))
carpeta2.agregar(carpeta5)

carpeta2.agregar(Archivo("Archivo B2"))

carpeta5 = Carpeta("Carpeta BB")
carpeta5.agregar(Archivo("Archivo BB1"))
carpeta5.agregar(Archivo("Archivo BB2"))
carpeta5.agregar(Archivo("Archivo BB3"))
carpeta5.agregar(Archivo("Archivo BB4"))
carpeta2.agregar(carpeta5)


raiz.agregar(carpeta2)

raiz.agregar(Archivo("Archivo 3"))
raiz.agregar(Archivo("Archivo 4"))

carpeta1 = Carpeta("Carpeta C")
carpeta1.agregar(Archivo("Archivo C1"))
carpeta1.agregar(Archivo("Archivo C2"))
carpeta1.agregar(Archivo("Archivo C3"))
carpeta1.agregar(Archivo("Archivo C4"))
raiz.agregar(carpeta1)


raiz.agregar(Archivo("Archivo 5"))


raiz.imprimaPreorden()
```
